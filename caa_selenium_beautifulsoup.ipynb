```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import selenium.common.exceptions as exceptions
import pandas as pd
```

```python
gdit_urls = ["https://www.caa.com/entertainmenttalent/touring/search#comedy",
        "https://www.caa.com/entertainmenttalent/touring/search#electronic"]
```

```python
length = len(gdit_urls)
for i in range(length):
    gdit_url = gdit_urls[i]
    global driver
    options = webdriver.ChromeOptions()
    options.add_argument("start-maximized")
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option("detach", True)
    options.add_experimental_option('useAutomationExtension', False)
    try:
        #driver = webdriver.Chrome(chrome_options=options, executable_path=r'C:\WEBDRIVERS\chromedriver.exe')
        driver = webdriver.Chrome()
        driver.get(gdit_url) #opens websites in the for loop
        time.sleep(7) #dynamically loading the website.... 


        previous_height = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") #scrolls to the bottom of the page

        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") #execute scrolling

            time.sleep(7)

            new_height = driver.execute_script("return document.body.scrollHeight") #height will change because page is longer
            if new_height == previous_height:
                break
            previous_height = new_height

        page_source = driver.page_source #downloads all the html

        soup = BeautifulSoup(page_source, 'html.parser')

        artists = []

        for span in soup.find_all('span', attrs = {"class": "artist-name"}):
            artists.append(span.text.strip())

        cleaned_artists = [artist.replace(",", "").replace("'", "") for artist in artists]

        # Display each name on a new line
        result = "\n".join(cleaned_artists)
        print(result)
        driver.close()

    except exceptions.WebDriverException as e:
        print("you need to update the webdriver")
```
